```python
# Q1(a)
import numpy as np
def my_bisection(f, x,mu, a, b, tol , max_iterations=1000):
    """
    Approximates a root of the function f within the interval [a, b]
    to within a specified tolerance |f(a+b/2)| < tol.

    Args:
        f: The function for which we want to find the root.
        a, b: Interval endpoints.
        tol: Tolerance (desired accuracy).

    Returns:
        Approximated root.
    """
    # Check if a and b bound a root
    # if f(a) * f(b) >= 0:
    #     raise Exception("The scalars a and b do not bound a root")
    i = 0
    x_k = x

    while np.any(abs((b-a) / 2) >= tol):
        direction = d(x,mu)
    # for i in range(max_iterations):
        m = (a + b) / 2  # Midpoint
        x_k = x + m * direction
        output = f(x_k,mu,direction)
        print(f'Iteration {i}: a={a}, b={b}, m={m}, theta_prime(m)={output}, x_k={x_k}')

        if output > 0:
            b = m
        else:
            a = m
        i += 1
        if i==max_iterations:
            break
    return (a + b) / 2

mu = 1/2
x = [4,2]
a = -100.0
b = 100.0
# Define the objective function and constraints
def objective_function(x):
  return x[0]**2 + 4*x[1]**2 - 8*x[0] - 16*x[1]

def constraint1(x):
  return x[0] + x[1] - 5

def constraint2(x):
  return x[0] -3

def d(x, mu):
    if max(0,constraint1(x)) >0:
        # return -np.array([2*x[0] - 8 + 2*mu*(x[0] + x[1] - 5), 8*x[1] - 16 + 2*mu*(x[0] + x[1] - 5) + 2*mu*(x[0]-3)])
        return -np.array([2*x[0] - 8 + 2*mu*(x[0] + x[1] - 5), 8*x[1] - 16 + 2*mu*(x[0] + x[1] - 5) ])
    else:
        return -np.array([2*x[0] - 8, 8*x[1] - 16])

# Define the augmented Lagrangian function
theta = lambda x, l, mu: objective_function(x + l * d(x, mu)) + mu * max(0, constraint1(x + l * d(x, mu))) ** 2  \
         +mu * max(0, constraint2(x + l * d(x, mu))) ** 2
def theta_prime(x,mu,dr):
    if max(0,constraint1(x)) >0:
        # return 2*x[0]*dr[0] + 8*x[1]*dr[1] - 8*dr[0]- 16*dr[1] + 2*mu*(dr[0] + dr[1])*constraint1(x)+ 2*mu * dr[0] * constraint2(x)
        return 2*x[0]*dr[0] + 8*x[1]*dr[1] - 8*dr[0]- 16*dr[1] + 2*mu*(dr[0] + dr[1])*constraint1(x)
    else:
        return 2*x[0]*dr[0] + 8*x[1]*dr[1] - 8*dr[0]- 16*dr[1]

# f_prime = lambda x, l, mu: -d(x + l*d(x,mu), mu)

print(f'The initial d: {d(x,mu)}')

# when x=(4,2), the initial lambda will be 1/(7)
# with given tolerance .00001 and [a,b] = [-100, 100], checked
my_bisection(theta_prime, x,mu, a, b, tol=.00001)
```
