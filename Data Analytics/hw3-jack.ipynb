$$\begin{aligned}
SS_R &= \sum (\hat{y}- \bar{y})^2 \\
&= \sum (\hat{y}^2- 2\hat{y}\bar{y}+ \bar{y}^2) \\
&= \sum\hat{y}^2 - 2\bar{y} \sum \hat{y} +n\bar{y}^2 \\
\end{aligned}$$

$\because$

$$\begin{aligned}
\sum\hat{y} &= \sum(\hat{\beta}_0 +\hat{\beta}_1x ) \\
&= n\hat{\beta}_0 + n\bar{x}\hat{\beta}_1 \\
&= n(\hat{\beta}_0 + \bar{x}\hat{\beta}_1) \\
&= n\bar{y}
\end{aligned}$$

$\therefore$

$$\begin{aligned}
SS_R &= \sum\hat{y}^2 - 2\bar{y} \sum \hat{y} +n\bar{y}^2 \\
&= \sum\hat{y}^2 - n\bar{y}^2
\end{aligned}$$

```python

# importing pandas
import pandas as pd
import numpy as np
import seaborn as sns
from sklearn.linear_model import LinearRegression
```

```python
# Read Text Files with Pandas

col_names = ['mpg','cylinders','displacement','horsepower',
            'weight','acceleration','year','origin','car_name']
# read text file into pandas DataFrame
df = pd.read_fwf("auto-mpg.data.txt",header=None,names = col_names)
# df = pd.DataFrame(df1.to_numpy() , columns=col_names)
# display DataFrame
print(df)
corr = df.corr()
# plot the heatmap

cmap = cmap=sns.diverging_palette(6, 250, as_cmap=True)

def magnify():
    return [dict(selector="th",
                 props=[("font-size", "14pt")]),
            dict(selector="td",
                 props=[('padding', "0em 0em")]),
            dict(selector="th:hover",
                 props=[("font-size", "20pt")]),
            dict(selector="tr:hover td:hover",
                 props=[('max-width', '400px'),
                        ('font-size', '20pt')])
]

corr.style.background_gradient(cmap, axis=1)\
    .set_properties(**{'max-width': '80px', 'font-size': '10pt'})\
    .set_caption("Car data correlation matrix")\
    .format(precision=2)\
    .set_table_styles(magnify()).set_table_styles(magnify())
```

```python
# find to missing value
df[df.isin({'?'}).any(1)]
```

There are 6 missing horsepower values. After cleanning data, the result looks identical to lecture slides.

```python
df = df[~df.isin({'?'}).any(1)]
corr = df.corr()
print('Check the data dimention is ', df.shape)
corr.style.background_gradient(cmap, axis=1)\
    .set_properties(**{'max-width': '80px', 'font-size': '10pt'})\
    .set_caption("Car data correlation matrix")\
    .format(precision=2)\
    .set_table_styles(magnify()).set_table_styles(magnify())


    
```

```python
col_x = df.columns[1:-1]
df_model = df[col_x]

for i in col_x:
    
    y = df_model[i]
    # list true in current index
    #temp = lambda a,l:[True if x == a else False for x in l]
    x = df_model.drop([i],axis=1)
    model = LinearRegression().fit(x,y)
    r_sq = model.score(x, y)
    print('when y is', i, ', R-square: ', r_sq)
    print('VIF:', 1/(1-r_sq))
    # print('beta_0 :', model.intercept_)
    # print('All other betas: ', model.coef_)

```
